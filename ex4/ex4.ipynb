```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from skimage import exposure

def plot_histogram(image, title, ax):
    # Convert image to uint8 if it's not already
    if image.dtype != np.uint8:
        image = (image * 255).astype(np.uint8)
    hist = cv2.calcHist([image], [0], None, [256], [0, 256])
    ax.plot(hist)
    ax.set_xlim([0, 256])
    ax.set_title(title)

def match_histograms(source, reference):
    # Match the histogram of the source image to that of the reference image
    return exposure.match_histograms(source, reference, channel_axis=None)

# Load images
dark_image = cv2.imread('dark.jpg', cv2.IMREAD_GRAYSCALE)
light_image = cv2.imread('bright.jpg', cv2.IMREAD_GRAYSCALE)
low_contrast_image = cv2.imread('low_contrast.png', cv2.IMREAD_GRAYSCALE)
high_contrast_image = cv2.imread('flower.jpeg', cv2.IMREAD_GRAYSCALE)

# Target histogram: Using light image as the reference
target_histogram = light_image

# Apply histogram matching
matched_dark = match_histograms(dark_image, target_histogram)
matched_low_contrast = match_histograms(low_contrast_image, target_histogram)
matched_high_contrast = match_histograms(high_contrast_image, target_histogram)

# Plot original and matched images with their histograms
fig, axes = plt.subplots(4, 4, figsize=(20, 20))

# Dark Image
axes[0, 0].imshow(dark_image, cmap='gray')
axes[0, 0].set_title('Original Dark Image')
axes[0, 0].axis('off')
plot_histogram(dark_image, 'Histogram of Dark Image', axes[0, 1])

axes[0, 2].imshow(matched_dark, cmap='gray')
axes[0, 2].set_title('Matched Dark Image')
axes[0, 2].axis('off')
plot_histogram(matched_dark, 'Histogram of Matched Dark Image', axes[0, 3])

# Light Image
axes[1, 0].imshow(light_image, cmap='gray')
axes[1, 0].set_title('Original Light Image')
axes[1, 0].axis('off')
plot_histogram(light_image, 'Histogram of Light Image', axes[1, 1])

# Low Contrast Image
axes[2, 0].imshow(low_contrast_image, cmap='gray')
axes[2, 0].set_title('Original Low Contrast Image')
axes[2, 0].axis('off')
plot_histogram(low_contrast_image, 'Histogram of Low Contrast Image', axes[2, 1])

axes[2, 2].imshow(matched_low_contrast, cmap='gray')
axes[2, 2].set_title('Matched Low Contrast Image')
axes[2, 2].axis('off')
plot_histogram(matched_low_contrast, 'Histogram of Matched Low Contrast Image', axes[2, 3])

# High Contrast Image
axes[3, 0].imshow(high_contrast_image, cmap='gray')
axes[3, 0].set_title('Original High Contrast Image')
axes[3, 0].axis('off')
plot_histogram(high_contrast_image, 'Histogram of High Contrast Image', axes[3, 1])

axes[3, 2].imshow(matched_high_contrast, cmap='gray')
axes[3, 2].set_title('Matched High Contrast Image')
axes[3, 2].axis('off')
plot_histogram(matched_high_contrast, 'Histogram of Matched High Contrast Image', axes[3, 3])

# Clear out the remaining subplots
for i in range(4):
    for j in range(2, 4):
        axes[i, j].axis('off')

# Adjust layout to remove white space
plt.tight_layout()
plt.savefig('output_histogram_matching.png', bbox_inches='tight', pad_inches=0)
plt.show()
```
