```python
from PIL import Image
import matplotlib.pyplot as plt

# Load the image
image_path = "bubble.jpeg"
image = Image.open(image_path)

# Resize the image using different interpolation methods
resized_images = {}

# Shrinking the image to half its size
resized_images['Nearest Neighbor (Shrink)'] = image.resize(
    (image.width // 2, image.height // 2), Image.NEAREST)
resized_images['Bilinear (Shrink)'] = image.resize(
    (image.width // 2, image.height // 2), Image.BILINEAR)
resized_images['Bicubic (Shrink)'] = image.resize(
    (image.width // 2, image.height // 2), Image.BICUBIC)

# Zooming the image to twice its size
resized_images['Nearest Neighbor (Zoom)'] = image.resize(
    (image.width * 2, image.height * 2), Image.NEAREST)
resized_images['Bilinear (Zoom)'] = image.resize(
    (image.width * 2, image.height * 2), Image.BILINEAR)
resized_images['Bicubic (Zoom)'] = image.resize(
    (image.width * 2, image.height * 2), Image.BICUBIC)

# Plotting the results
fig, axs = plt.subplots(2, 3, figsize=(15, 10))

for i, (title, img) in enumerate(resized_images.items()):
    ax = axs[i // 3, i % 3]
    ax.imshow(img)
    ax.set_title(title)
    ax.axis('off')

plt.tight_layout()
plt.show()
```
