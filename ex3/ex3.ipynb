```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load the image
image_path = 'star.jpg'
image = cv2.imread(image_path)

# Convert the image to RGB (OpenCV loads images in BGR format by default)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the original image
plt.imshow(image_rgb)
plt.title('Original Image')
plt.show()

# Sharpen the image
sharpening_kernel = np.array([[-1, -1, -1],
                              [-1,  9, -1],
                              [-1, -1, -1]])
sharpened_image = cv2.filter2D(image_rgb, -1, sharpening_kernel)

# Convert to grayscale and apply thresholding
gray_image = cv2.cvtColor(sharpened_image, cv2.COLOR_RGB2GRAY)
_, thresholded_image = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)

# Adjust brightness and contrast
def adjust_brightness_contrast(image, brightness=0, contrast=1.0):
    adjusted_image = cv2.convertScaleAbs(image, alpha=contrast, beta=brightness)
    return adjusted_image

brightness = 50
contrast = 1.2
adjusted_image = adjust_brightness_contrast(sharpened_image, brightness, contrast)

# Display the results
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

axes[0].imshow(sharpened_image)
axes[0].set_title('Sharpened Image')
axes[0].axis('off')

axes[1].imshow(thresholded_image, cmap='gray')
axes[1].set_title('Thresholded Image')
axes[1].axis('off')

axes[2].imshow(adjusted_image)
axes[2].set_title('Brightness and Contrast Adjusted Image')
axes[2].axis('off')

plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load the image
image_path = 'star.jpg'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Display the original image
plt.figure(figsize=(12, 4))
plt.subplot(1, 4, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Simple Thresholding
_, simple_thresh = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)

# Adaptive Thresholding
adaptive_thresh_mean = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_MEAN_C,
                                             cv2.THRESH_BINARY, 11, 2)
adaptive_thresh_gaussian = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                                 cv2.THRESH_BINARY, 11, 2)

# Otsu's Thresholding
_, otsu_thresh = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# Display the results
plt.subplot(1, 4, 2)
plt.imshow(simple_thresh, cmap='gray')
plt.title('Simple Thresholding')
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(adaptive_thresh_mean, cmap='gray')
plt.title('Adaptive Mean Thresholding')
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(adaptive_thresh_gaussian, cmap='gray')
plt.title('Adaptive Gaussian Thresholding')
plt.axis('off')

plt.figure(figsize=(6, 4))
plt.imshow(otsu_thresh, cmap='gray')
plt.title("Otsu's Thresholding")
plt.axis('off')

plt.show()
```
